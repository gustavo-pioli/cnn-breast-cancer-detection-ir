```python
import os
import cv2

# Diretórios de entrada e saída
input_dir = "/mnt/d/Documentos Gustavo/TCC/processed_heatmaps"
output_dir = "/mnt/d/Documentos Gustavo/TCC/processed_heatmaps3"

# Criar output_dir se não existir
os.makedirs(output_dir, exist_ok=True)

# Percorrer todas as subpastas
for root, _, files in os.walk(input_dir):
    for file in files:
        partes = file.split(".")  # Divide o nome do arquivo por "."

        # Verifica se tem pelo menos 3 partes e se está nas imagens desejadas
        if len(partes) > 2 and partes[2] in ["1", "3", "5"]:
            # Verifica se não termina com "F" ou "Z"
            if not (file.endswith("F.jpg") or file.endswith("Z.jpg")):
                # Caminho completo do arquivo de origem
                caminho_arquivo = os.path.join(root, file)

                # Ler imagem com cv2
                imagem = cv2.imread(caminho_arquivo)

                if imagem is not None:
                    # Criar o mesmo caminho relativo no diretório de saída
                    caminho_relativo = os.path.relpath(caminho_arquivo, input_dir)
                    destino_arquivo = os.path.join(output_dir, caminho_relativo)

                    # Criar diretório se não existir
                    os.makedirs(os.path.dirname(destino_arquivo), exist_ok=True)

                    # Salvar imagem no novo diretório usando cv2
                    cv2.imwrite(destino_arquivo, imagem)
                    print(f"✅ Copiado: {caminho_arquivo} ➝ {destino_arquivo}")
                else:
                    print(f"⚠️ Erro ao carregar: {caminho_arquivo}")
```

```python
import os

# Diretório de saída onde estão as imagens filtradas
output_dir = "/mnt/d/Documentos Gustavo/TCC/processed_heatmaps3"

# Percorrer as pastas Sick e Healthy
for classe in ["Sick", "Healthy"]:
    pasta_classe = os.path.join(output_dir, classe)
    
    if not os.path.exists(pasta_classe):
        continue  # Pula se a pasta não existir

    # Lista de pacientes dentro da pasta Sick ou Healthy
    pacientes = os.listdir(pasta_classe)

    for paciente in pacientes:
        pasta_paciente = os.path.join(pasta_classe, paciente)

        if os.path.isdir(pasta_paciente):  # Confirma se é um diretório
            imagens = [f for f in os.listdir(pasta_paciente) if f.endswith(".jpg")]

            # Dicionário para armazenar a primeira imagem encontrada de cada tipo
            imagens_por_tipo = {"1": None, "3": None, "5": None}

            for imagem in imagens:
                partes = imagem.split(".")
                
                if len(partes) > 2 and partes[2] in imagens_por_tipo:
                    tipo = partes[2]

                    if imagens_por_tipo[tipo] is None:
                        # Guarda a primeira imagem encontrada desse tipo
                        imagens_por_tipo[tipo] = imagem
                    else:
                        # Remove imagens excedentes desse tipo
                        caminho_excedente = os.path.join(pasta_paciente, imagem)
                        os.remove(caminho_excedente)
                        print(f"🗑️ Removido: {caminho_excedente}")
```

```python
import os
import cv2
import numpy as np
import random

# Diretório onde estão os pacientes doentes
output_dir = "/mnt/d/Documentos Gustavo/TCC/processed_heatmaps3"
pasta_sick = os.path.join(output_dir, "Sick")

# Encontrar o maior ID existente
existentes = [d for d in os.listdir(pasta_sick) if d.startswith("p") and d[1:].isdigit()]
if existentes:
    max_id = max(int(d[1:]) for d in existentes)  # Maior ID numérico
else:
    max_id = 0  # Se não houver pacientes, começa do zero

# Novo ID começa de (último ID + 1000)
novo_paciente_id = max_id + 1000  

# Função para aplicar Data Augmentation
def augment_image(image):
    # Rotação aleatória entre -10 e 10 graus
    angle = random.uniform(-10, 10)
    h, w = image.shape[:2]
    M = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1)
    rotated = cv2.warpAffine(image, M, (w, h))

    # Espelhamento aleatório
    if random.choice([True, False]):
        rotated = cv2.flip(rotated, 1)  # Espelhar horizontalmente

    # Ajuste de brilho e contraste aleatório
    alpha = random.uniform(0.8, 1.2)  # Contraste
    beta = random.randint(-20, 20)    # Brilho
    brightened = cv2.convertScaleAbs(rotated, alpha=alpha, beta=beta)

    # Adicionar ruído gaussiano aleatório
    noise = np.random.normal(0, 10, brightened.shape).astype(np.uint8)
    noisy = cv2.add(brightened, noise)

    return noisy

# Criar novos pacientes a partir dos existentes
for paciente in os.listdir(pasta_sick):
    pasta_paciente = os.path.join(pasta_sick, paciente)

    if os.path.isdir(pasta_paciente):  # Confirma se é um diretório
        imagens = [f for f in os.listdir(pasta_paciente) if f.endswith(".jpg")]

        # Criar 2 novos pacientes por cada original
        for _ in range(2):
            novo_paciente = f"p{novo_paciente_id}"
            pasta_novo_paciente = os.path.join(pasta_sick, novo_paciente)
            os.makedirs(pasta_novo_paciente, exist_ok=True)

            for imagem in imagens:
                caminho_imagem = os.path.join(pasta_paciente, imagem)
                img = cv2.imread(caminho_imagem)

                if img is not None:
                    img_aug = augment_image(img)
                    destino = os.path.join(pasta_novo_paciente, imagem)
                    cv2.imwrite(destino, img_aug)

            print(f"✅ Novo paciente criado: {novo_paciente}")
            novo_paciente_id += 1  # Incrementa o ID para o próximo paciente

print("📢 Data augmentation finalizado!")
```
